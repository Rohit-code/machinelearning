```python
import pandas as pd
import numpy as np
movies = pd.read_csv('input/tmdb_5000_movies.csv')
credits = pd.read_csv('input/tmdb_5000_credits.csv')
movies=movies.merge(credits,on='title')
#now we use only some for recommendition
# movie_id,title,release_date,genres,keywords,overview,cast,crew
movies=movies[['movie_id','title','release_date','genres','overview','keywords','cast','crew']]
movies.head(1)
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
movies.isnull().sum()
```

```python
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
movies['keywords']=movies['keywords'].apply(convert)
movies['cast']=movies['cast'].apply(convert)
movies.head(1)
```

```python
import ast
def director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append('name')
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(director)
movies['overview'][0]
#making overview column split in list
movies['overview']=movies['overview'].apply(lambda x:x.split())
movies.head()
```

```python
#removig spaces in words
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies['overview']=movies['overview'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies.head(1)
```

```python
new_file=movies[['movie_id','title','genres','release_date','overview',]]
new_file['overview']=new_file['overview'].apply(lambda x:" ".join(x))
new_file
```

```python
new_file['overview'][0]
```

```python
new_file['overview']=new_file['overview'].apply(lambda x:x.lower())
```

```python
new_file.head(1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features=5000,stop_words='english')
vectors=cv.fit_transform(new_file['overview']).toarray()
vectors
```

```python
cv.get_feature_names()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
stem('in the 22nd century, a paraplegic marine is dispatched to the moon pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization.')
```

```python
new_file['overview']=new_file['overview'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
similarity[0]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])
```

```python
def recommend(movie):
    movie_index=new_file[new_file['title']==movie].index[0]
    distance=similarity[movie_index]
    movies_list=sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:10]
    for i in movies_list:
        print(new_file.iloc[i[0]].title)
    return
```
